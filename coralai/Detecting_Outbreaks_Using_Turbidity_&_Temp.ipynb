```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# temperature in F
x2 = np.array([86, 63, 74, 90, 55, 80, 76, 84, 100, 110, 82])

# turbidity
x1 = np.array([8.6, 8.9, 8.2, 9.0, 9.2, 7.5, 6.0, 5.5, 10.1, 11.2, 7.8])

#incredibly obivious connection here ^^

# Combining the two temperature variables into a 2D array
X = np.column_stack((x1, x2))

# danger of outbreaks
y = np.array([1, 1, 0, 1, 1, 0, 0, 0, 1, 1, 0])
```

```python
plt.scatter(x2, y, label='Temperature')
plt.scatter(x1, y, label='Turbidity', marker='x')
plt.title("Cause of Outbreaks")
plt.xlabel("Temperature in F & Turbidity Levels")
plt.ylabel("Chance of Outbreak(1: Yes, 0: No)")
plt.legend()
plt.show()
```

```python
# c parameter prevents overfitting, and implements regularization
# solver parameters implements optimization
# multi_class parameter
logreg = LogisticRegression(C=1.0, solver="lbfgs", multi_class="ovr")

# already in 2D array which is required format


logreg.fit(X, y)
```

```python
# Predict the outcome for a specific input
new_data = np.array([[9.6, 80]])  # New temperature and turbidity values
predicted_outcome = logreg.predict(new_data)
print("Predicted outcome:", predicted_outcome)

# ISSUE HERE: Not enough data / training to determine what to
# do when temp is high, but turbidity is low or other way around.
```
